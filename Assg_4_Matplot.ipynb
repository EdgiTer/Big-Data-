```python
plt.hist(df)
```

```python
plt.plot(df['total_profit'])
plt.xlabel(['month_number'])
plt.ylabel(['total_units'])
plt.show()
```

```python
plt.plot(df['bathingsoap'])
plt.plot(df['facecream'])
plt.plot(df['moisturizer'])
plt.plot(df['facewash'])
plt.plot(df['shampoo'])
plt.plot(df['toothpaste'])
plt.xlabel(['month_number'])
plt.ylabel(['total_units'])
plt.show()
```
