```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
sy=sum(y)
sx=sum(x)
n=len(x)
sx2=list(map(lambda num:num**2,x))
sx2=sum(sx2)
sxy=list(map(lambda a,b:a*b,x,y))
sxy=sum(sxy)
print(sy,sx,sx2,sxy)

a=(sy*sx2-sx*sxy)/(n*sx2-sx*sx)
b=(n*sxy-sx*sy)/(n*sx2-sx*sx)
print(a,b)

ypd=list(map(lambda w:a+b*w,x))
plt.scatter(x, y)
plt.plot(x, ypd,c='red')
plt.show()
```
